# Series

It is defined as a one dimensional array that is capable of storing various data types.

```python
# Example
import pandas as pd
```

```python
data = [2,5,64,34,22,1,4,5]
a = pd.Series(data)
print(a)
print(type(a))
```

```python
# Getting the specific value from series
print(a[4])
```

```python
# Changing the index and data type of data
b = pd.Series(data, index = ["a", "b", "c", "d", "e", "f", "g", "h"], dtype="float", name = "sampleData") # name parammeter is used to define the name of data
print(b)
```

```python
# Working series with dictionaries
dic = {"Name": ["Python", "C++", "Java", "C"], "Rank": [1, 3, 2, 4]}
dic_series = pd.Series(dic)
print(dic_series)
```

```python
sample_series_example = pd.Series(12, index=[1,3,4,5,6])
print(sample_series_example)
```

```python
# Merging two data 
first_series = pd.Series(12, index=[1,2,3,4,5,6,7])
second_series = pd.Series(12, index=[1,2,3,4])
print(first_series + second_series)
```

# DataFrame

A DataFrame is a data structure that organizes data into a 2-dimensional table of rows and columns, much like a spreadsheet

```python
#dataframe using list
list_one = [1,3,4,5,6,7,8]
result = pd.DataFrame(list_one)
print(result)
print(type(result))
```

```python
# dataframe using dictionary
example_two = {"Col1": [1,3,4,5,6,7], "Col2": [10,11,12,13,56, 70]}
example_two_result = pd.DataFrame(example_two)

# fetching only col1 from the above 
example_two_result = pd.DataFrame(example_two, columns=["Col1"])
print(example_two_result)
```

```python
# Combining two series to generate a dataframe
sr = {"s": pd.Series([1,2,3,45]), "r": pd.Series([7,8,9,10])}
result = pd.DataFrame(sr)
print(result)
print(type(result))
```

# Handeling Missing Values

# Dropna

The dropna() method removes the rows that contains NULL values.

```python
parkinson_csv_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Parkinsson disease.csv")
```

```python
parkinson_csv_data.isna().sum()
```

```python
# parkinson_csv_data.dropna(axis=0) #(axis = 1 => drops the null value via column) and  (axis = 0 => drops the null value via row)

# parkinson_csv_data.dropna(how="all") #(how = "any" => drops all the rows with nan values) and (how = "all" => drops the only row containing nan values)

# dropping the nan values in a specific column via subset parameter
# parkinson_csv_data.dropna(subset=["MDVP:Jitter(%)"])

# This is because function with inplace=True does not return anything. It modifies an existing data frame with the desired operation and does it 'in place', that is on the original data frame.
parkinson_csv_data.dropna(inplace=True)
print(parkinson_csv_data)
```

# GroupBy

The groupby is one of the most frequently used Pandas functions in data analysis. It is used for grouping the data points (i.e. rows) based on the distinct values in the given column or columns.

```python
grp_by_example = pd.DataFrame({"Name": ["C", "C++", "Java", "JS", "Python", "C", "C", "C++"], 
                               "C1": [10, 20, 10, 20, 30, 20, 10,20],
                              "C2": [80,60,60,60,30,60,50, 80]})
print(grp_by_example)
```

```python
result = grp_by_example.groupby("Name")
for x, y in result:
    print(x)
    print(y)
```

# Melt

Pandas melt() function is used to change the DataFrame format from wide to long. It is used in reshaping of data

```python
melt_example = pd.DataFrame({"days": [1,2,3,4,5,6], "eng": [10,11,12,13,14,15], "maths": [21,22,23,24,25,26]})
print(melt_example)
```

```python
# reshaping the above data using melt function
pd.melt(melt_example, id_vars=["days"])
```
